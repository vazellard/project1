```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver
```

```python
driver = webdriver.Chrome('chromedriver')
driver.get('https://lewan.baidu.com/rankland?idfrom=5056&gameSource=standalone')
time.sleep(3)
driver.find_element_by_xpath('//*[@id="root"]/div/div/div[2]/div/div/div[1]/div[2]/div[1]/div[1]/div/div/div[5]').click()
time.sleep(3)
game_list = []
num = 0
genre_chi = ['独立游戏', '动作', '角色扮演', '策略经营', '休闲', '射击', '益智', '赛车', '体育', '桌游棋牌', '音舞', '其它', '策略经营', '生存', '模拟', '沙盒', '街机', '竞技', '模拟战争', '武侠', '火影忍者', 
             '牌桌', '冒险', '格斗游戏', '回合制', '卡牌',  '塔防']
genre_kor = ['인디', '액션', '롤플레잉', '전략경영', '캐주얼 퍼즐', 'FPS', '두뇌게임', '레이싱', '스포츠', '보드', '리듬', '기타', '전략', '서바이벌', '시뮬레이션', '샌드박스', '아케이드', '경쟁', '전쟁 시뮬레이션', '무협', '나루토'
             '카드', '어드벤처', 'FTG파이트게임', 'mmorpg', '하스스톤', '타워디펜스게임']

for i in range(20):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    a_list = soup.find_all(class_ = 'info-box')
    for item in a_list:
        num += 1
        title = item.find(class_="cname").text
        genre = item.find(class_="type").text.split("/")
        for ge in genre:
            if ge in genre_chi:
                genre[genre.index(ge)] = genre_kor[genre_chi.index(ge)]
        rank = num
        game_list.append([rank, title, genre])
    driver.find_element_by_css_selector('#root > div > div > div.layout-main > div > div > div.main-area > div.top-area-box > div.top-area > div > div.type-rank-main > div.right > ul > li.rc-pagination-next').click()  
    time.sleep(2)
    
game_list
```

```python
frame = ['순위', '제목', '장르']
df = pd.DataFrame(game_list, columns=frame)
df
```

```python
maxlen = 0
game_list2 = []
for item in game_list:
    if len(item[2]) > maxlen:
        maxlen = len(item[2])
for game in game_list:
    info = []
    info.append(game[0])
    info.append(game[1])
    info.append(game[2])
    for genre in game[3]:
        if genre not in info:
            info.append(genre)
    game_list2.append(info)
for list in game_list2:
    if len(list) < 3+maxlen :
        for i in range(3+maxlen - len(list)):
            list.append(None)
    
```

```python
frame2 = ['순위', '제목']
for i in range(maxlen):
    frame2.append('장르'+str(i+1))
df = pd.DataFrame(game_list2, columns=frame2)
df.to_csv('project/mobile_chi_data.csv', index=False)
```
